LOOP CONTROL STATMENTS

```python
#break,continue,pass
'''break (The Emergency Exit)
The break statement stops the loop entirely and jumps to the code immediately following the loop. 
Use it when you've found what you’re looking for and don’t need to keep searching.

Logic: "Stop right now and get out.
'''
'''continue (The Skip Button)
The continue statement skips the rest of the code inside the current loop iteration
and jumps straight back to the top for the next round.

Logic: "Skip this specific one and keep going.'''

'''pass (The Placeholder)
The pass statement does absolutely nothing. In Python, you cannot have an empty block of code
(like an if statement or a function with no content).
pass is a "null operation" used to prevent syntax errors.'''
```

```python
# break

for i in range(1,11):
    if i == 6:
        break
    print(i)
```

```python
for i in range(1,11):
    if i == 8:
        break  #in real time traffic light time frame //cnn etc
    print(i)
```

```python
for i in range(1,11):
    if i == 8:#skips 8
        continue#time traffic light time frame //cnn etc
    print(i)
```

WHILE LOOP

```python
'''\\The while Loop
A while loop repeats as long as a certain condition is True.
It is "indefinite," meaning it could theoretically run forever if the condition never becomes False.'''
```

```python
print('datascience')# one time is ok
```

```python
print('datascience')# c & p ok for 5 times
print('datascience')
print('datascience')
print('datascience')
print('datascience')
```

```python
i = 1 #initializing

while i<=5:#condition
    print('datascience')# for better expirence try in pycharm
    i = i + 1 #increment
              # if we dont give increment it will loop infinite
    
```

```python
i = 5 #initializing

while i>=1:#condition
    print('datascience:',i)# for better expirence try in pycharm
    i = i - 1 #decrement
              
```

```python
'''can we use multiple while loops
which means nested while loop
to understand better use pycharm debug with f8 option'''

# WHILE LOOP IS USED WHEN WE HAVE CONDITIONS.
```

```python
i = 1

while i<=5:
    print("ds")#outer loop prints one time
    j = 1
    while j<=4:
        print("os")#inner loop prints four times then go to outer loop..>repeat
        j = j + 1
    i = i + 1
    
    print()
```

```python
i = 1

while i<=5:
    print("ds", end = " ")#outer loop prints one time
    j = 1
    while j<=4:
        print("os", end = " ")#inner loop prints four times then go to outer loop..>repeat
        j = j + 1
    i = i + 1
    
    print()
```

```python
i = 1

while i <= 2:
    j = 0 
    while j <= 2:
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

```python
i = 1

while i <= 4:
    j = 0 
    while j <= 3:
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```
