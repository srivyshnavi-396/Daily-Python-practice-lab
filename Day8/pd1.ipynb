but it should be in blw 30 and 40 as there is one point.the reason is that the point is OUTLIER
OUTLIER----it is something which is totally differ from remaining group(the numbers which are very large or very small than other numbers in dataset it is called outlier).
these outlier impact on machine learning algorithms.
to handle tis outliers we need to understand probability....
it should not be removed but should be adjusted.(we can adjust using logistic regession and naivebais algorithm)
to adjust outlier we use sigmoid function(probability function)
detecting outlier is also called as anomoly detection.
we detect outlier using graphs.

```python
vis5 = sns.lmplot(data = df, x="InternetUsers", y="BirthRate")
plt.show(vis5)
```

```python
vis6 = sns.lmplot(data = df, x="InternetUsers", y="BirthRate", fit_reg=False)#line will disappear
plt.show(vis6)
```

```python
vis7 = sns.lmplot(data = df, x="InternetUsers", y="BirthRate", fit_reg=True)#line will appear
plt.show(vis7)
```

```python
vis8 = sns.lmplot(data = df, x="InternetUsers", y="BirthRate", fit_reg= False ,hue = "IncomeGroup")#hue - parameter for color
plt.show(vis8)
```

by the color we can read the graph easy way.
highincome group has lowest internet users and low income group has lowest.
