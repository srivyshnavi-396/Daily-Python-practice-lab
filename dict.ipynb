DICTIONARY

```python
'''A dictionary is one of the most powerful and flexible built in datatypes.
..>it is unordered(but preserved by insertion order since python 3.7)
..>it is a mutable collection of Key-Value pairs.
..>think of it is like a real life dictionary: you look up for a specific word(key)
to find its defination(value).
..>A python dictionary is a collection of key and value pairs separated by a colon (:) & enclosed 
in curly braces {}.
..>Keys must be unique in a dictionary, duplicate values are allowed.'''
```

Copying Dict

```python
dict4 = {'fruits':'apple','veg':'carrot','drinks':'juice'}
dict4
```

```python
dict5 = dict4
```

```python
dict5#reference of dict4
```

```python
id(dict4),id(dict5)#same ids
```

```python
dict6 = dict5.copy()#creates a copy of dict5
```

```python
dict6
```

```python
id(dict5)
```

```python
id(dict6)#diff ids
```

```python
dict4['course'] = ['cse']#dict5 will also impact as it is pointing to dict4(original)
print(dict5)
print(dict4)
```

ALL & ANY 

```python
'''The all() method returns: 
True - If all all keys of the dictionary are true 
False - If any key of the dictionary is false 
The any() function returns True if any key of the dictionary is True. If not, any() returns False. '''
```

```python
dict8 = {1:0,2:3}
all(dict8)
```

```python
dict8 = {0:1,2:3}
all(dict8)
```

```python
dict8 = {0:1,0:3}
any(dict8)
```

```python
dict8 = {1:0,2:3}
any(dict8)
```
