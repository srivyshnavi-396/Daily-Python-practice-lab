'''1. The try Block
Definition: The try block is a container for code that has the potential to raise an error. It does not handle errors; it simply monitors them.

Logic: As soon as a line in the try block fails, Python stops executing the rest of the try block and jumps immediately to the except block.

Syntax: try:
           #code

2. The except Block
Definition: This is the "Handler." It defines the logic that should run if a specific exception occurs. It prevents the program from terminating.

Logic: You can have multiple except blocks for different types of errors (e.g., one for FileNotFoundError and one for PermissionError).

Syntax: except:
             #code


3. The else Block
Definition: The else block contains code that should execute only if no exceptions were raised in the try block.

Logic: This is useful for code that should only run if the "risky" operation was successful. It helps keep the try block clean by separating the risky code from the "success" code.

Syntax: else:
           #code

4. The finally Block
Definition: The finally block is the "Clean-up" block. It executes regardless of whether an exception occurred or not.

Logic: Even if your program crashes or uses a return statement, the finally block will still run. It is essential for releasing external resources like database connections or network sockets.

Syntax: finally:
             #code


5. Raising Exceptions (raise)
Definition: The raise keyword is used to manually trigger an exception when a specific condition (defined by the programmer) is met.

Logic: This is used for "Business Logic" errorsâ€”situations where the code is technically valid but the data is wrong for your specific app.
    
Syntax: raise #eg:valueerror()
'''

'''Why use raise?
While Python automatically raises errors for things like dividing by zero, you use raise to:

Enforce Business Logic: e.g., rejecting a negative age or an invalid username.

Rethrow Errors: Catching an error, logging it, and then letting it continue up the chain.

Custom Errors: Triggering specific exceptions you've defined yourself.'''

```python
'''1. Raising a Built-in Exception
The most common use is checking a condition and triggering a standard error like ValueError or TypeError.'''
def set_age(age):
    if age < 0:
        # Manually triggering an error with a custom message
        raise ValueError("Age cannot be negative!")
    print(f"Age set to {age}")

try:
    set_age(-5)
except ValueError as e:
    print(f"Caught an error: {e}")
```

```python
'''2. Re-raising an Exception
Sometimes you want to catch an error just to perform an action (like logging), but you still want the program to eventually stop or handle it elsewhere.'''
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Log: Someone tried to divide by zero again.")
    raise  # This sends the original ZeroDivisionError back up the stack
```

```python
'''3. Exception Chaining (raise ... from)
If you want to raise a new error because another one occurred, you can use from to show the "trace" of how you got there.'''
try:
    import non_existent_module
except ImportError as e:
    raise RuntimeError("Required library is missing") from e
```
