TUPLE OPERATIONS

```python
# even though we cannot change the tuple we can perform following operations.
#JOINING(CONCATENATION)
#you can combine two tuples to create a brand-new third tuple.
t1 = (1,2)
t2 = (3,4)
t3 = t1 + t2
t3
```

```python
#PACKING
fruits = ('apple','grapes')
fruits
```

```python
#UNPACKING 
#we can also extract tuple values in to variables in one line.

fruits = ('apple','grapes')
'''should be variable but not strings because it will no allocate to literal like string'''
(green,blue)= fruits
print('blue')
```

```python
#REPEATING it is all about repeating a tuple using * operator.
t = (('hi',)*3)
t
```

```python

fruits = ('apple','grapes')
(green,blue) = fruits
green
```
