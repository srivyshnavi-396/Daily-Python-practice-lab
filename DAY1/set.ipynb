SETS

```python
'''A SET IS AN UNORDERED COLLECTION OF UNIQUE ELEMENTS.
..>SETS ARE UNORDERED COLLECTION OF DATA AS THERE IS NO INDEX BASED POSITION.
..>THEY ARE UNINDEXED COLLECTION OF ITEMS ,YOU CANNOT ACCESS THEM VIA INDEX.
..>SET ELEMENTS ARE UNIQUE , NO DUPLICATE ELEMENTS ARE ALLOWED TO IT.
..>SET ELEMENTS ARE IMMUTABLE ,WHICH CANNOT BE CHANGED.
..>SET ITSELF IS MUTABLE, WE CAN ADD OR REMOVE ITEMS FROM IT.
....>SETS ARE HIGHLY OPTIMIZED FOR MEMBERSHIP TESTING AND TO PERFORM MATHEMATICAL OPERATIONS LIKE UNIONS AND INTERSECTIONS.'''
```

'''WHY MUTABLE AND IMMUTABLE
....>SET ITSELF IS MUTABLE ...WHICH MEANS THE ITEMS IN THE SET CAN BE CHANGEABLE.
....>SET ELEMENTS SHOULD BE IMMUTABLE LIKE TUPLE,STRING,AND INTEGERS..ETC
BUT IT WONT ALLOW LIST AS IT IS MUTABLE.
...>PYTHON INSISTS THAT EVERY ELEMENT YOU PUT IN TO SET MUST BE UNCHANGABLE(IMMUTABLE).'''

....>DUPLICATES ARE NOT ALLOWED
....>SET ITSELF IS MUTABLE BUT ITEMS IN IT MUST BE IMMUTABLE.

SET OPERATIONS

```python
#UNION
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {9,10}
A|B#union of a and b(all elements from both sets,no duplicates)
```

```python
A.union(B)#union of a and b(all elements from both sets,no duplicates)
```

```python
A.union(B,C)#union of a,b and c
```

```python
""" 
Updates the set calling the update() method with union of A , B & C. 
For below example Set A will be updated with union of A,B & C."""
A.update(B,C)
A
```
