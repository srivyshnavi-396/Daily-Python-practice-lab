LIST CREATION USING CONSTRUCTOR

```python
# in general we use square braces to create a list.
#but we can also create a list using constructor()
'''the list constructor list() is built in fuction used to create a new list object
the constructor is particularly useful when you need to convert other data type in to list.
syntax: l = list(iterable)
the iterable can be any sequence like (string or tuple) 
or a collection like (set or dict) or any object you can loop over.

Constructor is just like funtion call
it is used to convert other data types in to list
used for typeconversion.'''
```

```python
#use cases
#coverting a tuple or set in to list
#why to convert....tuples are immutable and sets are unordered..
#so converting them in to list can allows you to modify or sort the data.
```

```python
#as it will consider the above declared variable as list so, i will delete list
del list
```

```python
#converting tuple in to list
my_tuple = (10,20,30)
my_list = list(my_tuple)
print(my_list)
```

```python
#converting set into list
set = {1,2,3}
my_lst = list(set)
print(my_lst)
```

```python
#converting a string in to list
'''when you pass string in to list() constructor, 
it breaks the string in to individual characters.'''
text = "python"
lst = list(text)
print(lst)
```

```python
#creating a list from range
#this is the most frequent uses of the constructor,
# especially when you need a sequence of numbers for loop or calculation.
numbers = list(range(1,6))
print(numbers)
```

```python
#converting Dictionary keys and values
#if you want to list keys or values from dict.
dict = {"fruit" : "apple", "season":"winter","month":"nov"}
keys = list(dict)#to list keys we use same syntax
print(keys)
```

```python
dict = {"fruit" : "apple", "season":"winter","month":"nov"}
values = list(dict.values())#to list values we must use variablename.(dot)values
print(values)
```

```python
#calling list with no arguments creates an empty list.
empty = list()
print(empty)
```

ALL/ANY IN LIST 

```python
#all() method  returns:
True - if all elements in a list are true.
False - if any element in list is false.
#the any() function returns true if any element in the list is true.
#if not, any() returns false.
```

```python
lst1 = [1,2,3,4,0]#will return false if atleast one value is false.
all(lst1)
```

```python
any(lst1)# will return true as if atleast one value is true.
```

```python
lst2 = [1,2,3,4,True,False]
all(lst2)
```

```python
any(lst2)
```

```python
lst2 = [1,2,3,4,True]# return true if all values are true.
all(lst2)
```

```python
lst2 = [-1,-2,-3,0,True,False]
all(lst2)
```

```python
lst2 = [-1,-2,-3,0,True,False]#returns true if atleast one value is true.
any(lst2)
```

```python
lst2 = [-1,-2,-3,0,False]
any(lst2)
```
